```python
from pandas_datareader import data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Select stocks and time range for analysis
stocks     = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'ORCL']
start_date = '2012-06-01'
end_date   = '2018-12-01' 

# load yahoo data for analysis
st_data = data.DataReader(stocks,'yahoo',start_date,end_date)
st_data.head()
```

```python
close_price = st_data['Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
close_price = close_price.reindex(all_weekdays).fillna(method='ffill')
close_price.describe()
```

```python
avg_return = close_price.diff().mean()
std_var    = close_price.diff().std()
corr       = close_price.diff().corr()
rf         = 0.02/365
print (avg_return, std_var, corr)
```
